```python
def trans_c_s(x,y,z):
    r = np.sqrt(x**2 + y**2 + z**2)
    #print r
    cociente = z/r
    if abs(cociente) > 1:
        #print "cociente mayor a 1"
        if cociente < 0:
            theta = np.arccos(-1)
            #print "cociente negativo", theta
        else:
            theta = np.arccos(1)
            #print theta
    else:
        
        theta = np.arccos(z/r)
        #print theta
    #Aqui hay un problema relevante: cada vez que y o x sean nulos, habra un problema
    #de indefinicion de operacion
    if x == 0.:
        #print "x=0"
        if y == 0.:
            #print "y=0"
            phi = 2*np.pi*np.random.rand()
        else:
            if y > 0:
                phi = np.pi/2
            else:
                phi = -np.pi/2
    else:
        
        if x < 0:
            #print "x<0"
            phi = np.arctan(y/x) + np.pi
        else:
            #print "x debería ser positivo"
            if y < 0:
                #print "y<0 tambien!"
                phi = np.arctan(y/x) + 2*np.pi
            else:
                #print "x<0 pero y>0"
                phi = np.arctan(y/x)
            
    return r, theta, phi

```
